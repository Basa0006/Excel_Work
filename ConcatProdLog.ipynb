```python
#Create a function that takes the five manufacturing locations and makes them one column with the manufacturing location



def conditions(mfgc, mfgi, mfgm, mfgb, mfgd):
    if mfgc == "X":
        return "EGV"
    elif mfgi == "X":
        return "Algona"
    elif mfgm == "X":
        return "Mexico"
    elif mfgb == "X":
        return "Brazil"
    elif mfgd == "X":
        return "Denmark"
    else:
        return "WTF"
    
    
```

```python
#Call conditions function created above to fill Manufactured column with manufacturing location
df1['Manufactured']=""
df1['Manufactured'] = df1.apply(lambda x: conditions(x['MFG Chicago'], x['MFG Iowa'], x['MFG Mexico'], x['MFG B'], x['MFG O.1']), axis=1)
df1
df1.info()
```
